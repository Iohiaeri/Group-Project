```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
colnames=['day', 'time', 'tide']
df = pd.read_fwf(r'ASTR19_S22_group_project_data.txt', header=None, skiprows=(0,1,2), names=colnames)
x = np.linspace(1, 82, 82)
y = df['tide']
y_err = np.full(82,.25)
yfit = 2.5*-np.cos(np.pi*x)+2.48
```

```python
plt.errorbar(x,y,y_err,fmt='o')
plt.plot(x, yfit)
plt.xlabel("Number of Waves Charted")
plt.ylabel("Wave Height")
plt.show()
```

```python
ydata = df['tide'].tolist()
yfitdata = yfit.tolist()
```

```python
residual = []
for i in range(82):
    res = (ydata[i]-yfit[i])
    if res > 0:
        residual.append(res-.25)
    elif res < 0:
        residual.append(res+.25)
    else:
        residual.append(res)
```

```python
plt.hist(residual, bins=45)
plt.xlabel("Resdiual Amount")
plt.ylabel("Resdiual Frequency")
plt.savefig("Histo.pdf")
print(f"The standard Deviation is {np.std(residual)}")
```
